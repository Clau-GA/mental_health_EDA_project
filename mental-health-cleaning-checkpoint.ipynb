```python
# See all unique values in the "no_employees" column
df["no_employees"].unique()
```

```python
# Fix corrupted company size values

df["no_employees"] = df["no_employees"].replace({
    "5-Jan": "1-5",          # Fix Excel's mistaken date conversion
    "25-Jun": "6-25"         # Fix Excel's mistaken date conversion
})

# Check again to confirm the correction
df["no_employees"].unique()
```

```python
# Show all unique gender entries to see how messy the data is
df["Gender"].unique()
```

```python
# STEP 4B — Standardize the gender column (lowercase + remove spaces)

df["Gender_clean"] = (
    df["Gender"]
    .str.strip()        # Remove spaces at start/end
    .str.lower()        # Lowercase everything
    .str.replace(r"[^a-zA-Z\- ]", "", regex=True)  # Remove symbols like ? ^ _ etc.
)
```

```python
df["Gender_clean"].unique()
```

```python
# STEP 4E — Map all messy gender entries into clean categories

# 1) Define a mapping from cleaned raw values → grouped categories
gender_map = {
    # --- Male variants ---
    "male": "male",
    "m": "male",
    "man": "male",
    "male cis": "male",
    "cis male": "male",
    "cis man": "male",
    "msle": "male",
    "maile": "male",
    "mail": "male",
    "mal": "male",
    "malr": "male",
    "make": "male",
    "male-ish": "male",
    "guy -ish ": "male",  # as it appears in Gender_clean
    "something kinda male": "male",
    "ostensibly male unsure what that really means": "male",
    "male leaning androgynous": "male",

    # --- Female variants ---
    "female": "female",
    "f": "female",
    "woman": "female",
    "female cis": "female",
    "cis female": "female",
    "cis-femalefemme": "female",
    "femake": "female",
    "femail": "female",
    "female trans": "female",
    "trans-female": "female",
    "trans woman": "female",

    # --- Nonbinary / gender-diverse ---
    "non-binary": "nonbinary",
    "enby": "nonbinary",
    "genderqueer": "nonbinary",
    "androgyne": "nonbinary",
    "agender": "nonbinary",
    "fluid": "nonbinary",
    "neuter": "nonbinary",

    # --- Queer labels ---
    "queer": "queer",
    "queershethey": "queer",

    # --- Other / unclear / joke entries ---
    "all": "other",
    "nah": "other",
    "a little about you": "other",
    "p": "other"
}

# 2) Apply the mapping to create a new final gender column
df["Gender_final"] = df["Gender_clean"].map(gender_map)  # Map known values

# 3) Any value not found in the map becomes NaN, so we label it as "other"
df["Gender_final"] = df["Gender_final"].fillna("other")

# 4) Check the final categories
df["Gender_final"].value_counts()
```

```python
# Check the range of ages
df["Age"].describe()
```

```python
# Sort ages to see the lowest and highest values
df["Age"].sort_values().head(15)

# Look at the highest values
df["Age"].sort_values().tail(15)
```

```python
# STEP 4H — Clean the Age column

# 1) Convert to numeric (force errors to NaN)
df["Age_clean"] = pd.to_numeric(df["Age"], errors="coerce")

# 2) Replace unrealistic ages with NaN
df.loc[(df["Age_clean"] < 18) | (df["Age_clean"] > 100), "Age_clean"] = None

# 3) Check the new age distribution
df["Age_clean"].describe()
```
