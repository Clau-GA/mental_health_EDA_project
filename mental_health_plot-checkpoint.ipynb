```python
# Get counts of each treatment response (Yes / No)
treatment_counts = df["treatment"].value_counts()

treatment_counts
```

```python
# MATPLOTLIB CHART 1: Overall Treatment Count

# Create a new figure and set its size
plt.figure(figsize=(6, 4))

# x positions (0, 1) for the bars
x_positions = range(len(treatment_counts))

# Create the bar chart
plt.bar(
    x_positions,                     # x positions
    treatment_counts.values          # heights (counts)
)

# Set the labels for each bar (No, Yes)
plt.xticks(
    x_positions,                     # positions
    treatment_counts.index           # category labels
)

# Add title and axis labels
plt.title("Treatment Count (Yes vs No) - Matplotlib")
plt.xlabel("Treatment Response")
plt.ylabel("Number of Respondents")

plt.tight_layout()
plt.show()
```

```python
# Create a pivot table: rows = Gender_group, columns = treatment, values = count
gender_treatment = df.pivot_table(
    index="Gender_group",
    columns="treatment",
    aggfunc="size",
    fill_value=0
)

gender_treatment
```

```python
# MATPLOTLIB CHART 2: Grouped bars for Gender vs Treatment

plt.figure(figsize=(10, 6))

# x positions for each gender group
x = range(len(gender_treatment))

# bar width (how fat each bar is)
bar_width = 0.35

# Bars for "No"
plt.bar(
    [pos - bar_width/2 for pos in x],           # shift left
    gender_treatment['No'],
    width=bar_width,
    label="No"
)

# Bars for "Yes"
plt.bar(
    [pos + bar_width/2 for pos in x],           # shift right
    gender_treatment['Yes'],
    width=bar_width,
    label="Yes"
)

# X-axis labels
plt.xticks(
    x,
    gender_treatment.index,
    rotation=45
)

# Labels and title
plt.title("Treatment by Gender Group - Matplotlib")
plt.xlabel("Gender Group")
plt.ylabel("Number of Respondents")

plt.legend()
plt.tight_layout()
plt.show()
```
